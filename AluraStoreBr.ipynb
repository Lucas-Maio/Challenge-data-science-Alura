### Importação dos dados



```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
faturamento = loja['Preço'].sum()
print(faturamento)

faturamento = loja2['Preço'].sum()
print(faturamento)

faturamento = loja3['Preço'].sum()
print(faturamento)

faturamento = loja4['Preço'].sum()
print(faturamento)
```

```python
faturamento_loja = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
valores = [faturamento_loja, faturamento_loja2, faturamento_loja3, faturamento_loja4]

plt.figure(figsize=(8, 5))
plt.bar(lojas, valores, color='seagreen')
plt.title('Faturamento por Loja')
plt.xlabel('Loja')
plt.ylabel('Faturamento (R$)')
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

# 2. Vendas por Categoria


```python
venda_categoria = loja.groupby('Categoria do Produto')['Preço'].sum()
print(f"Loja 1 \n\n{venda_categoria}\n")

venda_categoria = loja2.groupby('Categoria do Produto')['Preço'].sum()
print(f"Loja 2 \n\n{venda_categoria}\n")

venda_categoria = loja3.groupby('Categoria do Produto')['Preço'].sum()
print(f"Loja 3 \n\n{venda_categoria}\n")

venda_categoria = loja4.groupby('Categoria do Produto')['Preço'].sum()
print(f"Loja 4 \n\n{venda_categoria}\n")
```

# 3. Média de Avaliação das Lojas

```python
media_avaliacoes = loja['Avaliação da compra'].mean()
print(f"Loja 1 \n\n{media_avaliacoes}\n")

media_avaliacoes = loja2['Avaliação da compra'].mean()
print(f"Loja 2 \n\n{media_avaliacoes}\n")

media_avaliacoes = loja3['Avaliação da compra'].mean()
print(f"Loja 3 \n\n{media_avaliacoes}\n")

media_avaliacoes = loja4['Avaliação da compra'].mean()
print(f"Loja 4 \n\n{media_avaliacoes}\n")
```

```python
media_loja = loja['Avaliação da compra'].mean()
media_loja2 = loja2['Avaliação da compra'].mean()
media_loja3 = loja3['Avaliação da compra'].mean()
media_loja4 = loja4['Avaliação da compra'].mean()

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
medias = [media_loja, media_loja2, media_loja3, media_loja4]

plt.figure(figsize=(6, 6))
plt.pie(medias, labels=lojas, autopct='%1.2f%%', startangle=90, colors=['#66b3ff','#99ff99','#ffcc99','#ff9999'])
plt.title('Participação na Média das Avaliações por Loja')
plt.axis('equal')
plt.show()
```

# 4. Produtos Mais e Menos Vendidos

```python
produtos_mais_vendidos = loja['Produto'].value_counts()
print(f"Loja 1 \n\n{produtos_mais_vendidos}\n")

produtos_mais_vendidos = loja2['Produto'].value_counts()
print(f"Loja 2 \n\n{produtos_mais_vendidos}\n")

produtos_mais_vendidos = loja3['Produto'].value_counts()
print(f"Loja 3 \n\n{produtos_mais_vendidos}\n")

produtos_mais_vendidos = loja4['Produto'].value_counts()
print(f"Loja 4 \n\n{produtos_mais_vendidos}\n")

```

# 5. Frete Médio por Loja


```python
frete = loja['Frete'].mean()
print(f"Loja 1 \n\n{frete}\n")

frete = loja2['Frete'].mean()
print(f"Loja 2 \n\n{frete}\n")

frete = loja3['Frete'].mean()
print(f"Loja 3 \n\n{frete}\n")

frete = loja4['Frete'].mean()
print(f"Loja 4 \n\n{frete}\n")
```

```python
frete_loja = loja['Frete'].sum()
frete_loja2 = loja2['Frete'].sum()
frete_loja3 = loja3['Frete'].sum()
frete_loja4 = loja4['Frete'].sum()

lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
fretes = [frete_loja, frete_loja2, frete_loja3, frete_loja4]

plt.figure(figsize=(8, 5))
plt.plot(lojas, fretes, marker='o', linestyle='-', color='darkorange', linewidth=2)
plt.title('Frete Total por Loja')
plt.xlabel('Loja')
plt.ylabel('Frete Total (R$)')
plt.grid(True, linestyle='--', alpha=0.6)


for i, valor in enumerate(fretes):
    plt.text(i, valor + 5, f'R${valor:.2f}', ha='center')

plt.tight_layout()
plt.show()
```

## 📊 Relatório Final – Análise Comparativa de Lojas

### 🧭 Introdução

O presente relatório tem como objetivo auxiliar o Senhor João a decidir qual de suas quatro lojas virtuais deve ser vendida. Para isso, foram realizadas análises comparativas com base em dados de vendas, avaliações de clientes, fretes, categorias de produtos e faturamento total. Com esses dados, buscamos identificar a loja que apresenta o melhor desempenho geral e a que representa menor valor estratégico para o negócio.

---

### 📈 Desenvolvimento da Análise

#### 1. **Faturamento Total**

Analisando a soma dos preços dos produtos vendidos:

* **Loja 1** apresentou o maior faturamento, sendo a mais lucrativa do grupo.
* **Loja 4** teve o menor volume de vendas em valores financeiros.

Essa métrica é essencial, pois reflete o retorno financeiro gerado por cada loja.

#### 2. **Categorias de Produtos Mais e Menos Vendidas**

Observando o volume de vendas por categoria:

* As **categorias mais vendidas** em geral foram: **eletrodomésticos** e **móveis**.
* A **categoria com menor volume de vendas** foi **brinquedos**, com baixa representatividade.

Esse fator ajuda a compreender o apelo comercial dos produtos de cada loja.

#### 3. **Média das Avaliações dos Clientes**

As notas de avaliação dos clientes foram fundamentais para medir a satisfação:

* **Loja 1** obteve a melhor média de avaliações, demonstrando boa experiência de compra.
* **Loja 4** teve a menor média, indicando possíveis problemas operacionais ou de atendimento.

Boas avaliações influenciam diretamente na fidelização de clientes e na reputação da loja.

#### 4. **Produtos Mais e Menos Vendidos**

Com base na contagem de vendas:

* Produtos como **"Assistente virtual"** e **"Micro-ondas"** foram os mais vendidos.
* Produtos da categoria **brinquedos** foram os menos populares.

A análise ajuda a entender o comportamento dos consumidores e quais produtos trazem mais retorno.

#### 5. **Frete Médio**

Comparando o custo médio do frete por loja:

* **Loja 3** apresentou o **menor frete médio**, o que pode atrair mais compradores.
* **Loja 2** teve o frete mais elevado, o que pode impactar negativamente nas vendas.

Fretes mais baixos aumentam a competitividade e a taxa de conversão no e-commerce.

---

### ✅ Conclusão e Recomendação

Após a análise criteriosa de todos os dados, recomenda-se que o **Senhor João venda a Loja 4**.

#### **Justificativa:**

* **Menor faturamento** entre as quatro lojas.
* **Pior média de avaliação dos clientes**, indicando baixa satisfação.
* **Desempenho inferior nas categorias mais populares**, como eletrodomésticos e móveis.
* Apesar do frete estar dentro de uma média aceitável, a loja **não apresenta diferenciais competitivos** em nenhuma das métricas analisadas.

Por outro lado, **a Loja 1 é a mais estratégica**, com **melhor faturamento e avaliações**, sendo **altamente recomendada para manter no portfólio**.
